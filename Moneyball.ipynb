```python
positions = m_positions[['playerID', 'yearID', 'G_p', 'G_c', 'G_1b', 'G_2b', 'G_3b','G_ss', 'G_lf', 'G_cf', 'G_rf']]
positions.head()
```

```python
salaries = m_salaries[['playerID', 'yearID', 'salary']]
salaries.head()
```

```python
players = m_players[['playerID', 'nameFirst', 'nameLast']]
players.head()
```

```python
sal_player = pd.merge(salaries, players)
sal_player.head()
```

```python
t_batting = m_batting[['playerID', 'yearID', 'stint', 'teamID', 'AB', 'H', 'BB', 'HBP', 'SF']]
t_batting.head()
```

```python
sal_player_bat = pd.merge(sal_player, t_batting)
sal_player_bat.head()
```

```python
sal_player_bat['OBP'] = (sal_player_bat['H'] + sal_player_bat['BB'] + sal_player_bat['HBP']) \
                        / (sal_player_bat['AB'] + sal_player_bat['BB'] + sal_player_bat['HBP'] + sal_player_bat['SF'])
sal_player_bat.count()
```

```python
sal_player_bat['ratio'] = sal_player_bat['OBP'] / (sal_player_bat['salary']/100000)
sal_player_bat.head(10)
```

```python
sal_player_bat = sal_player_bat[sal_player_bat.salary > 0]
sal_player_bat = sal_player_bat[sal_player_bat.AB > 50]
sal_player_bat = sal_player_bat[sal_player_bat.OBP > 0]
sal_player_bat = sal_player_bat[sal_player_bat.OBP < 1]
sal_player_bat.head()
```

```python
sal_player_bat = sal_player_bat.sort('ratio', ascending=False)
```

```python
all_data = pd.merge(positions, sal_player_bat)
all_data.sort('ratio', ascending=False).head()
```
